# FuzzyWuzzy с cities.csv

```python
import csv
from fuzzywuzzy import fuzz, process
```

```python
#в кодировке Windows 1251
file_cities_csv = "cities.csv"
#в кодировке UTF-8
file_cities_utf8_csv = "cities_utf8.csv"
```

```python
search_text = "вороWW"

with open(file_cities_csv) as f:
    reader = csv.DictReader(f, delimiter = ';')
    for line in reader:
        print(f"{line['Город']} - {fuzz.WRatio(search_text, line['Город'])}")
        
```
