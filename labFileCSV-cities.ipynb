```python
# в кодировке Window-1251
file_cties_csv = "cities.csv"
# в кодировке UTF-8
file_cties_utf8_csv = "cities_utf8.csv"
```

## Чтение из CSV-файла

```python
with open(file_cties_csv) as f:
    reader = csv.reader(f, delimiter=';')
    for line in reader:
        print(line)
```

```python
with open(file_cties_utf8_csv, encoding='utf-8') as f:
    reader = csv.reader(f)
    for line in reader:
        print(line)
```

```python
with open(file_cties_csv) as f:
    reader = csv.reader(f, delimiter=';')
    for n, line in enumerate(reader):
        if n >= 5:
            break
        print(line)
```

```python
with open(file_cties_csv) as f:
    reader = csv.DictReader(f, delimiter=';')
    for line in reader:
        if "Крым" in line['Регион']:
            print(line)
```

## Запись в CSV-файл

```python
temp_header = ""
temp_data = list()

with open(file_cties_csv) as f:
    reader = csv.DictReader(f, delimiter=';')
    temp_header =reader.fieldnames
    for line in reader:
        if line['Регион'] == "Оренбургская область":
            temp_data.append(line)

print(temp_header)
for line in temp_data:
    print(line)
```

```python
with open(os.path.join(TEMP_DIR, "temp_cities.csv"), 'w') as f:
    writer = csv.DictWriter(f, fieldnames=temp_header, delimiter=';', lineterminator='\n')
    writer.writeheader()
    writer.writerows(temp_data)
```

```python
with open(os.path.join(TEMP_DIR, "temp_cities_utf-8.csv"), 'w', encoding='utf-8') as f:
    writer = csv.DictWriter(f, fieldnames=temp_header, delimiter=';', lineterminator='\n')
    writer.writeheader()
    writer.writerows(temp_data)
```
