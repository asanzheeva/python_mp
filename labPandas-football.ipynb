## Группировка данных

```python
df.Nationality.unique()
```

```python
group_1 = df[df.Nationality == 'Argentina'].Age
group_2 = df[df.Nationality == 'Brazil'].Age
group_3 = df[df.Nationality == 'Spain'].Age
group_4 = df[df.Nationality == 'England'].Age

df_new = pd.DataFrame([group_1.mean(), group_2.mean(), group_3.mean(), group_4.mean()],
    columns=['mean(Age)'],
    index=['Argentina','Brazil','Spain', 'England'])
df_new
```

```python
group_1 = df[df.Nationality == 'Argentina'].Age
group_2 = df[df.Nationality == 'Brazil'].Age
group_3 = df[df.Nationality == 'Spain'].Age
group_4 = df[df.Nationality == 'England'].Age

df_new = pd.DataFrame([group_1.mean(), group_2.mean(), group_3.mean(), group_4.mean()],
    columns=['mean(Age)'],
    index=['Argentina','Brazil','Spain', 'England'])
df_new
```

```python
df_new = pd.DataFrame({
    'Min' : [group_1.min(), group_2.min(), group_3.min(), group_4.min()],
    'Mean' :[group_1.mean(), group_2.mean(), group_3.mean(), group_4.mean()],
    'Max' : [group_1.max(), group_2.max(), group_3.max(), group_4.max()]
})
df_new
```

## Сводная таблица

```python
df['Club'].isin(['FC Barcelona','Real Madrid', 'Manchester United', 'Juventus'])
```

```python
df[df['Club'].isin(['FC Barcelona','Real Madrid', 'Manchester United', 'Juventus'])]
```

```python
# \ + Enter - перенос строки
pivot = df[df['Club'].isin(['FC Barcelona','Real Madrid', 'Manchester United', 'Juventus'])]\
    .pivot_table(values=['Wage'],
                index=['Age'],
                columns=['Club'],
                aggfunc='max') 
pivot     
```

```python
# \ + Enter - перенос строки
pivot = df[df['Club'].isin(['FC Barcelona','Real Madrid', 'Manchester United', 'Juventus'])]\
    .pivot_table(values=['Wage'],
                index=['Nationality'],
                columns=['Club'],
                aggfunc='sum',
                margins=True) # добавляем столбец Итого
pivot     
```

```python
# \ + Enter - перенос строки
pivot = df[df['Club'].isin(['FC Barcelona','Real Madrid', 'Manchester United', 'Juventus'])]\
    .pivot_table(values=['Wage'],
                index=['Nationality'],
                columns=['Club'],
                aggfunc='sum',
                margins=True, # добавляем столбец Итого
                fill_value=0) # поменяли Nan на 0
pivot     
```
