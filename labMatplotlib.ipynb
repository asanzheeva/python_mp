```python
fig = plt.figure()
ax = fig.add_subplot()
fig.set_facecolor('lightgreen')
ax.set_facecolor('lightcoral')

ax.set_title('Название графика', size=20, color='red')
ax.set_xlabel('X (ось абцис)')
ax.set_ylabel('X (ось ординат)')
ax.set_xlim([-100,100])
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.plot([0, 1, 2, 3, 4, 5], [2, 3, 10, 5, 7, 1])
ax.plot([0, 1, 2, 3, 4, 5], [0, 1, 3, 2, 4, 6], linewidth=5)
ax.scatter([0, 1, 5, 3], [6, 10, 3, 1])

plt.show()
```

```python
x = [0, 1, 2, 3, 4, 5]
y1 = [2, 3, 10, 5, 7, 1]
y2 = [0, 1, 3, 2, 4, 6]

x3 = [0, 1, 5, 3]
y3 = [6, 10, 3, 1]

fig = plt.figure()
ax = fig.add_subplot()

ax.plot(x, y1)
ax.plot(x, y2, linewidth=5)
ax.scatter(x3, y3, color='red', marker='*')

plt.show()

```

```python
fig = plt.figure()
ax = fig.add_subplot()

x = np.linspace(0, 1, 10)
y = np.sin(x)

ax.set_title('График Sin(x)')
ax.plot(x,y)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot()

x = np.linspace(0, 10, 100)
y = np.sin(x)

ax.set_title('График Sin(x)')
ax.plot(x,y)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot()

x = np.linspace(0, 10, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

ax.set_title('График Sin(x) и Cos(x)')
ax.plot(x,y_sin, label='sin')
ax.plot(x,y_cos, label='cos')
ax.legend()

plt.show()
```
